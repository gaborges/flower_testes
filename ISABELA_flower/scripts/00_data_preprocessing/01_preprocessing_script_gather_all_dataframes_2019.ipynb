# Data pre-processing

```python
allData = pd.DataFrame()
for indexFolder, folder in folders.iterrows():

    # check f the path exists (becouse kaggle removed the + character from file names)
    fileName = folder['folder']
    if(not(exists("../../Dataset_ECUADOR_2019/"+folder['folder']+"/"+fileName+".csv"))):
        fileName = folder['folder'].replace("+","")
    
    # data = pd.read_csv("../../Dataset_ECUADOR_2019/0Jf4TH9Zzse0Z1Jjh7SnTOe2MMzeSnFi7feTnkG6vgs/0Jf4TH9Zzse0Z1Jjh7SnTOe2MMzeSnFi7feTnkG6vgs.csv", delimiter=',', encoding="utf8")
    data1 = pd.read_csv("../../Dataset_ECUADOR_2019/"+folder['folder']+"/"+fileName+".csv",
                       delimiter=',', encoding="utf8")
    #data.to_csv("teste1.csv", sep=',', encoding='utf-8', index=False)
    data2 = pd.read_csv("../../Dataset_ECUADOR_2019/"+folder['folder']+"/Sleep_Detection_"+fileName+".csv",
                       delimiter=',', encoding="utf8")
    #dataFLabel = pd.read_csv("../Dataset_ECUADOR_2019/"+folder['folder']+"/sleep_form_"+folder['folder']+"_treated.csv",
    #               delimiter=',', encoding="utf8")
    dataFLabel = pd.read_csv("../data_2019_processed/form/sleep_form_"+folder['folder']+"_treated.csv",
                   delimiter=',', encoding="utf8")
    # merge data
    mergedDS = pd.merge(data1, data2, on='timestamp')
    print(folder['folder']+" "+str(mergedDS.shape[0]))
    # remove unecessary attributes not relevant to sleep detection: foregroundApp,wifi,bluetooth,ble, risk,suburb,municipality,city,town,county
    del mergedDS["foregroundApp"]
    #del mergedDS["wifi"]
    #del mergedDS["bluetooth"]
    del mergedDS["ble"]
    del mergedDS["risk"]
    del mergedDS["suburb"]
    del mergedDS["municipality"]
    del mergedDS["city"]
    del mergedDS["town"]
    del mergedDS["county"]
    del mergedDS["Unnamed: 0"]
    del mergedDS["activity_y"]
    # rename
    mergedDS.rename(columns = {'activity_x':'activity'}, inplace = True)
    #del mergedDS["Unnamed: 0_x"]
    #del mergedDS["Unnamed: 0_y"]
    # give the status of awake to all dataset
    mergedDS['class'] = 'awake'
    # convert label DF to date time
    dataFLabel['timestamp'] = dataFLabel['timestamp'].astype('datetime64[ns]')
    mergedDS['timestamp2'] = 0
    for indexFolder, content in mergedDS['timestamp'].items():
        temp = datetime.strptime(content,'%Y-%m-%dT%H:%M:%SZ')
        mergedDS['timestamp2'][indexFolder] = temp.timestamp()
        
    #dataFLabel['sleep_time'] = dataFLabel['sleep_time'].astype('datetime64[ns]')
    #dataFLabel['wakeup_time'] = dataFLabel['wakeup_time'].astype('datetime64[ns]')
    # convert merged DF timestamp to datetime
    mergedDS['timestamp'] = mergedDS['timestamp'].astype('datetime64[ns]')
    # time of the day
    #mergedDS['minutes_day'] = mergedDS['timestamp'].hour * 60 + mergedDS['timestamp'].minute
    mergedDS['minutes_day'] = mergedDS.timestamp.dt.hour * 60 + mergedDS.timestamp.dt.minute
    # initialize times
    mergedDS['sleep_time'] = 0
    mergedDS['wakeup_time'] = 0
    mergedDS['time_sleeping'] = 0
    # alter class to sleep based on sleep and wakeuptimes
    for index, labelSet in dataFLabel.iterrows():
        sleep_time = dataFLabel['sleep_time'][index]
        wakeup_time = dataFLabel['wakeup_time'][index]
        # replace value
        for indexT, contentT in mergedDS['timestamp2'].items():
            if(sleep_time <= contentT) and (contentT <= wakeup_time):
                mergedDS['class'][indexT] = 'asleep'
            if (contentT < wakeup_time and mergedDS['time_sleeping'][indexT] == 0) or dataFLabel.index[dataFLabel.shape[0]-1]==indexT:
                sleep_time_dt = datetime.fromtimestamp(sleep_time, tz = None)     
                wakeup_time_dt = datetime.fromtimestamp(wakeup_time, tz = None)        
                mergedDS['sleep_time'][indexT] = sleep_time_dt.hour * 60 + sleep_time_dt.minute 
                mergedDS['wakeup_time'][indexT] = wakeup_time_dt.hour * 60 + wakeup_time_dt.minute 
                mergedDS['time_sleeping'][indexT] = wakeup_time - sleep_time 
        #mergedDS['class'] = np.where((sleep_time <= mergedDS['timestamp2'])
        #                              and (mergedDS['timestamp2'] <= wakeup_time), 'asleep', mergedDS['class'])
        #mergedDS['sleep_time'] = sleep_time.hour * 60 + sleep_time.minute # in minutes
        #mergedDS['wakeup_time'] = wakeup_time.hour * 60 + wakeup_time.minute  # in minutes
        #mergedDS['time_sleeping'] = mergedDS['time_sleeping'] /60
    # remove temp information
    del mergedDS['sleep_time']
    del mergedDS['wakeup_time']
    del mergedDS['time_sleeping']
    mergedDS['timestamp_text'] = mergedDS['timestamp']
    mergedDS['timestamp'] = mergedDS['timestamp2']
    del mergedDS['timestamp2']
    # replace the place from the class
    temp = mergedDS['class'].copy()
    del mergedDS['class']
    mergedDS['class'] = temp
    # print the guy
    mergedDS.to_csv(outputFolders+"student_"+folder['folder']+".csv", sep=',', encoding='utf-8', index=False)
    #mergedDS.to_csv("dados2019/student_"+folder['folder']+".csv", sep=',', encoding='utf-8', index=False)
    print(Counter(mergedDS['class']))
    # merge the data set if has both classes
    if('asleep' in Counter(mergedDS['class'])):
        if(allData.empty):
            allData = mergedDS
        else:
            #allData = allData.append(mergedDS, ignore_index=True)
            allData = pd.concat([allData, mergedDS])
        
# print all
#allData.to_csv(outputFolders+"allData.csv", sep=',', encoding='utf-8', index=False)
#allData.to_csv("dados2019/allData.csv", sep=',', encoding='utf-8', index=False)
```
